```python
#import data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
data_sale = pd.read_csv('./data/data-w1-1.csv')
```

```python
# know the data
# data_sale.describe()          
data_sale
```

```python
display_column =  list(data_sale.columns[1:])
data_sale_funiture_np = data_sale.iloc[0,1:].to_numpy()
data_sale_funiture_np_serial = pd.Series(data_sale_funiture_np)
data_sale_funiture_np_serial =data_sale_funiture_np_serial.map(lambda x : x.lstrip('$').replace(',','.'))
data_sale_funiture_np_serial_numeric = pd.to_numeric(data_sale_funiture_np_serial)
data_sale_funiture_np_serial_numeric
plt.plot(display_column,data_sale_funiture_np_serial_numeric)
plt.show()
```

```python
labels = data_sale.columns[1:]
data_sale_np = data_sale.iloc[:].to_numpy()

data_sale_np_funiture_serial =pd.to_numeric(pd.Series(data_sale_np[0,1:]).map(lambda x : x.lstrip('$').replace(',','.')))
data_sale_np_office_supply_serial =pd.to_numeric(pd.Series(data_sale_np[1,1:]).map(lambda x : x.lstrip('$').replace(',','.')))
data_sale_np_technology_serial =pd.to_numeric(pd.Series(data_sale_np[2,1:]).map(lambda x : x.lstrip('$').replace(',','.')))


x  = np.arange(len(labels))
width = 0.2 # width of the bar
fig,ax = plt.subplots()
rects1 = ax.bar(x-width,data_sale_np_funiture_serial,width,label='funiture')
rects2 = ax.bar(x,data_sale_np_office_supply_serial,width,label='office supply')
rects3 = ax.bar(x+width,data_sale_np_technology_serial,width,label='technology')

ax.set_ylabel('revenue ($)')
ax.set_title('sale report')
ax.set_xticks(x,labels)
ax.legend()

ax.bar_label(rects1,padding=3)
ax.bar_label(rects2,padding=3)
ax.bar_label(rects3,padding=3)
fig.tight_layout()
plt.show()
```
